# Credit Risk Classification

```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
```

## 1. Loading the dataset

```python
dataset = pd.read_csv("data/CreditRisk.csv", sep=";", encoding="latin-1")
dataset.head()
```

## 2. Preparing the data

```python
def historico_to_numeric(x):
    if x == "Negativa":
        return 2
    if x == "Desconhecida":
        return 1
    if x == "Positiva":
        return 0

dataset["historico-num"] = dataset["historico-credito"].apply(historico_to_numeric)
dataset.head()
```

```python
dataset.replace(
    {
        "inadimplente": {"Sim": 1, "Nao": 0},
        "garantias": {"Sim": 1, "Nao": 0},
        "historico-credito": {"Negativa": 1, "Desconhecida": 0, "Positiva": 2},
    },
    inplace=True,
)

dataset.head()
```

## 3. Converting the remaining categorical fields

```python
dataset.replace(
    {
        "endividamento": {
            "> 30 %": 1,
            "< 30 %": 0,
            "? 30 %": 2,
        },
        "valor-solicitado": {
            "< 15000.00": 1,
            "> 15000 < 35000": 2,
            "> 35000.00": 0,
        },
    },
    inplace=True,
)

dataset.head()
```

## 4. Feature extraction

```python
features = dataset[["historico-credito", "garantias", "endividamento", "valor-solicitado"]]
classes = dataset["inadimplente"]

features.head()
```

## 5. Training the model

```python
model = MultinomialNB()
model.fit(features, classes)
```

## 6. Predicting a new example

```python
new_contract = pd.DataFrame(
    [[1, 0, 1, 2]],
    columns=["historico-credito", "garantias", "endividamento", "valor-solicitado"],
)

prediction = model.predict(new_contract)
print(prediction)
```

## 7. Checking training accuracy

```python
correct = 0
wrong = 0

for i in range(0, len(features)):
    new_contract = features.iloc[i]
    data = pd.DataFrame(
        [new_contract],
        columns=["historico-credito", "garantias", "endividamento", "valor-solicitado"],
    )
    prediction = model.predict(data)

    if prediction[0] == classes.iloc[i]:
        correct += 1
    else:
        wrong += 1

accuracy = correct / (correct + wrong)

print("correct:", correct)
print("wrong:", wrong)
print("accuracy:", accuracy)
```
